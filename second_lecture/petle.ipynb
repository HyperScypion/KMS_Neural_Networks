# Pętle

### 1. For

```python
# wypisanie wszystkich liczb od 1 do 10 odpowiednik for(int i=1; i < 11; i++) z C++
for i in range(1, 11, 1): # pierwszy argument początek, drugi koniec, trzeci skok
# for i in range(1, 11) <- skrócony zapis    
    print(i, end=' ')
    
print('\n\n')

for i in range(0, 60, 10):
    print(i, end=' ')
    
print('\n\n')

for i in range(10, -1, -1):
    print(i, end=' ')
    
print('\n\n')
    
# Gdy mamy listę (nie tylko) możemy iterować się po jej elementach
# Podobna do foreach z np. C#
l = [1, 2, 3, 4, 5]
for i in l:
    print(i, end=' ')
    
print('\n\n')
    
# Można to również zapisać tak
for i in range(len(l)):
    print(l[i], end=' ')
```

### 2. While

```python
# Pętla trwająca dopóki spełniony jest warunek
a = 0
while(a < 10):
    print(a, end=' ')
    a += 1
    
# Pętla nieskończona
while(True):
    print('KMS')
```

### 3. For Else

```python
for i in range(0, 10):
    print(i, end=' ')
else:
    print('nie', end=' ')
```
